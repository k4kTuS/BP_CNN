# CLAHE vs HE comparison

```python
# img = cv.imread('../MURA-v1.1/train/XR_FOREARM/patient00063/study1_positive/image1.png')
img = cv.imread('../MURA-v1.1/train/XR_SHOULDER/patient00032/study1_positive/image1.png')
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

clahe = cv.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))

# fig, ax = plt.subplots(1, 5, figsize=(18, 10))
fig, ax = plt.subplots(1, 3, figsize=(16, 8))
fig.tight_layout()

ax[0].imshow(cv.cvtColor(gray_img, cv.COLOR_BGR2RGB))
ax[0].set_title("Original", fontsize=18, y=-0.08)

ax[1].imshow(cv.cvtColor(cv.equalizeHist(gray_img), cv.COLOR_BGR2RGB))
ax[1].set_title("Histogram equalization", fontsize=18, y=-0.08)

ax[2].imshow(cv.cvtColor(clahe.apply(gray_img.astype('uint8')), cv.COLOR_BGR2RGB))
ax[2].set_title("CLAHE", fontsize=18, y=-0.08)

for axis in ax:
    axis.axis('off')
# plt.show()
plt.savefig('hist_clahe_comp.jpg')
```
